```python
from py2neo import Graph, Node, Relationship
import json
import os, sys
sys.path.append(os.path.dirname(os.path.dirname(os.path.abspath('__file__'))))
import settings # import username and pssword!
```

```python
url = 'http://localhost:7474'
username = settings.NEO4J_USERNAME
password = settings.NEO4J_PASSWORD
graph = Graph(url + '/db/data/', username=username, password=password)
```

```python
def load_data_from_file(file_name):
    with open('../data/{0}'.format(file_name)) as data_file:
        data = json.load(data_file)
        return data
            
students = load_data_from_file("studentBehaviorInfo_all.json")
video_data_processed = load_data_from_file("video_data.json")
```

```python
def create_user(member_id):
    user = Node('User', id = member_id)
    graph.merge(user)

def create_video(video_id, videoSpeed, wordLevel, total_tests):
    video = Node('Video', id = video_id, videoSpeed = videoSpeed, wordLevel = wordLevel, total_tests = total_tests)
    graph.merge(video)

def create_relation(watched_order, average, skipped_tests, member_id, video_id):
    rel = Relationship(Node('User', id = member_id), str(watched_order), Node('Video', id = video_id), average=average, skipped_tests=skipped_tests)
    graph.merge(rel)
```

```python
# print ("video", "videoSpeed", "wordLevel", "average", "skipped_tests", "total_tests")
i = 1535
for student in students[1534:]:
    member_id = student['memberId']
    create_user(member_id)
    watched_order = 1
    for video in student['chosenVideo']:
        videoSpeed = video_data_processed[str(video)]['videoSpeed']
        wordLevel = video_data_processed[str(video)]['wordLevel']
        total = 0
        amount = 0
        total_tests = 0
        skipped_tests = 0
        for score in student['listenScore']:
            if score['postId'] == video:
                total_tests += 1
                if score['score'] != -1:
                    amount += 1
                    total += score['score']
                else:
                    skipped_tests += 1
        if amount != 0:
            average = float(total)/amount
        else:
            average = -1
#         print (video, videoSpeed, wordLevel, average, skipped_tests, total_tests)
        create_video(video, videoSpeed, wordLevel, total_tests)
        create_relation(watched_order, average, skipped_tests, member_id, video)
        watched_order += 1
    print "Finish : " + str(i)
    i += 1
print "Finish all !!"
```
